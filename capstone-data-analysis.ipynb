## About US Accidents Dataset

This dataset encompasses car accident data from 49 states within the USA, collected from February 2016 to March 2023. The data is sourced through multiple APIs that stream traffic incident data from various entities, including the US and state departments of transportation, law enforcement agencies, traffic cameras, and traffic sensors. The dataset currently contains around 7.7 million records.

#### Data Collection

The dataset was gathered in real-time using several traffic APIs. It includes comprehensive accident information for the contiguous United States over the specified period.

#### Dataset Description

The dataset consists of 46 columns and 7,728,394 rows. Below is a brief description of each column:

- **ID**: Unique identifier for each accident record.
- **Source**: Origin of the raw accident data.
- **Severity**: Accident severity, ranging from 1 (least impact) to 4 (most significant impact).
- **Start_Time**: Local time when the accident began.
- **End_Time**: Local time when the accident's impact on traffic ended.
- **Start_Lat**: Latitude of the accident's starting point.
- **Start_Lng**: Longitude of the accident's starting point.
- **End_Lat**: Latitude of the accident's ending point.
- **End_Lng**: Longitude of the accident's ending point.
- **Distance(mi)**: Length of the road affected by the accident, in miles.
- **Description**: Human-provided description of the accident.
- **Street**: Street name where the accident occurred.
- **City**: City where the accident occurred.
- **County**: County where the accident occurred.
- **State**: State where the accident occurred.
- **Zipcode**: Zip code of the accident location.
- **Country**: Country of the accident location.
- **Timezone**: Timezone of the accident location (e.g., Eastern, Central).
- **Airport_Code**: Nearest airport-based weather station to the accident location.
- **Weather_Timestamp**: Timestamp of the weather observation (local time).
- **Temperature(F)**: Temperature in Fahrenheit.
- **Wind_Chill(F)**: Wind chill in Fahrenheit.
- **Humidity(%)**: Humidity percentage.
- **Pressure(in)**: Air pressure in inches.
- **Visibility(mi)**: Visibility in miles.
- **Wind_Direction**: Wind direction.
- **Wind_Speed(mph)**: Wind speed in miles per hour.
- **Precipitation(in)**: Precipitation amount in inches.
- **Weather_Condition**: Weather condition (e.g., rain, snow, thunderstorm, fog).
- **Amenity**: Presence of amenities near the accident location.
- **Bump**: Presence of speed bumps or humps nearby.
- **Crossing**: Presence of pedestrian crossings nearby.
- **Give_Way**: Presence of give-way signs nearby.
- **Junction**: Presence of junctions nearby.
- **No_Exit**: Presence of no-exit signs nearby.
- **Railway**: Presence of railways nearby.
- **Roundabout**: Presence of roundabouts nearby.
- **Station**: Presence of stations nearby.
- **Stop**: Presence of stop signs nearby.
- **Traffic_Calming**: Presence of traffic calming measures nearby.
- **Traffic_Signal**: Presence of traffic signals nearby.
- **Turning_Loop**: Presence of turning loops nearby.
- **Sunrise_Sunset**: Period of day based on sunrise and sunset.
- **Civil_Twilight**: Period of day based on civil twilight.
- **Nautical_Twilight**: Period of day based on nautical twilight.
- **Astronomical_Twilight**: Period of day based on astronomical twilight.

For more details about this dataset, please refer to the dataset's information page.

```python
# Creating a function for sanity check
def simple_sanity_check(df):
    print('-' * 70)
    print(f'No. of Rows: {df.shape[0]}        No. of Columns: {df.shape[1]}')
    print('-' * 70)
    
    data_profile = pd.DataFrame({
        'DataType': df.dtypes,
        'Missing Values': df.isnull().sum(),
        'Missing %': (df.isnull().mean() * 100).round(2),
        'Unique Values': df.nunique()
    })
    
    print(data_profile)
    print('-' * 70)

# Call the function with your DataFrame 'df'
simple_sanity_check(df)
```

##### Feature Transformation changing timezones to UTC for consistency


```python
# Convert "End_Time" to datetime objects with error handling
object_columns_df["End_Time"] = pd.to_datetime(object_columns_df["End_Time"], errors='coerce')

# Define a dictionary mapping timezones to UTC offsets (hours)
timezone_offsets = {
    'US/Eastern': -5,   # Eastern Time (ET) UTC offset
    'US/Pacific': -8,   # Pacific Time (PT) UTC offset
    'US/Central': -6,   # Central Time (CT) UTC offset
    'US/Mountain': -7,  # Mountain Time (MT) UTC offset
    # Add more timezone offsets as needed
}

# Use NumPy broadcasting to add the corresponding UTC offset to "End_Time"
object_columns_df["End_Time_UTC"] = object_columns_df["End_Time"] + pd.to_timedelta(object_columns_df["Timezone"].map(timezone_offsets), unit='h')

# Convert "Start_Time" to datetime objects
object_columns_df["Start_Time"] = pd.to_datetime(object_columns_df["Start_Time"], errors='coerce')

# Use NumPy broadcasting to add the corresponding UTC offset to "Start_Time"
object_columns_df["Start_Time_UTC"] = object_columns_df["Start_Time"] + pd.to_timedelta(object_columns_df["Timezone"].map(timezone_offsets), unit='h')
```

### Feature engineering

```python
def extract_datetime_components(df, column_name):
    
    # Extract datetime components
    df[f'{column_name}_year'] = df[column_name].dt.year.astype(float)
    df[f'{column_name}_month'] = df[column_name].dt.month.astype(float)
    df[f'{column_name}_day'] = df[column_name].dt.day.astype(float)
    df[f'{column_name}_hour'] = df[column_name].dt.hour.astype(float)
    df[f'{column_name}_minute'] = df[column_name].dt.minute.astype(float)
    df[f'{column_name}_second'] = df[column_name].dt.second.astype(float)
    
    return df
```
